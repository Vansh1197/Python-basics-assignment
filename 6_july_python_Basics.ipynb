**Student Report Card Generator**

```python
name = input("Enter name: ")
math = int(input("Math marks: "))
science = int(input("Science marks: "))
english = int(input("English marks: "))

total = math + science + english
percentage = total / 3

print("\n--- Report Card ---")
print("Name:", name)
print("Math:", math)
print("Science:", science)
print("English:", english)
print("Total:", total)
print("Percentage:", percentage)
```

```python
### Student report card
name=input('Enter your name: ')
age=int(input('Enter your age: '))
class_std=int(input('Enter your class: '))

English_marks=int(input('Enter your English marks: '))
Hindi_marks=int(input('Enter your Hindi marks: '))
Maths_marks=int(input('Enter your Maths marks: '))
Science_marks=int(input('Enter your Science marks: '))

Total_Marks = English_marks + Hindi_marks + Maths_marks + Science_marks
Percentage = (Total_Marks / 400) * 100

print(f'Name: {name}')
print(f'Age: {age}')
print(f'Class: {class_std}')
print(f'Total Marks: {Total_Marks}')
print(f'Percentage: {Percentage}')
```
